```python
import numpy as np

# code to generate bracelets
def bracelet(a,b):
  brace = np.array([a,b])
  end = False
  while end == False:
    brace = np.append(brace,[int((brace[-1]+brace[-2]) % 10)])
    if brace[0]==brace[-2] and brace[1]==brace[-1]:
      end = True

  # Need to remove the last two elements to allow connection (no overlap)
  brace = brace[:-2]
  #print(brace)
  return brace


def possible_lengths():

  # Create our lists for sequence length and canonical sequences
  bracelet_lengths = []
  canonical_bracelets = []

  # We have two degrees of freedom in our starting choice so iterate over all of these
  for i in range(10):
    for j in range(10):
      # Create the bracelet for i,j starting points
      B_ij = bracelet(i,j)

      # generate a list of possible lengths
      found = False
      for k in range(len(bracelet_lengths)):
        if len(B_ij) == bracelet_lengths[k]:
          found = True

      if found == False:
        bracelet_lengths.append(len(B_ij))
        canonical_bracelets.append(B_ij)

  return bracelet_lengths


def find_canonical():

  canonical_bracelets = []

  # We have two degrees of freedom in our starting choice so iterate over all of these
  for i in range(10):
    for j in range(10):
      B_ij = bracelet(i,j)
      found = False
      # we now want to check if this sequence is a cyclic permutation
      for l in range(len(canonical_bracelets)):
        # make sure that we check sequences are the same length
        if len(B_ij) == len(canonical_bracelets[l]):
          # this is the target bracelet (first to appear)
          canonical_bracelet = canonical_bracelets[l]
          # concatenate the two sequences together
          check = np.concatenate([canonical_bracelet,canonical_bracelet], axis = None)
          # now want to scan through check to see if B_ij lies inside check
          for m in range(len(check) - len(B_ij)):
            # found matching pair
            if B_ij[0] == check[m] and B_ij[1] == check[m+1]:
              # Check its the exact same sequence
              for n in range(len(B_ij)):
                if B_ij[n] != check[m+n]:
                  found = False

              found = True

      if found == False:
        canonical_bracelets.append(B_ij)

    return canonical_bracelets


# actually, since
L = possible_lengths()
print(L)

total = 0
for k in range(len(L)):
  total += L[k]

print(str(total) + "\n")


canon_brace = find_canonical()
for k in range(len(canon_brace)):
  print(len(canon_brace[k]))
  print(str(canon_brace[k]) + "\n")


# since these total to 100, there are 100 possible seqence types

#a = np.array([1,2])
#b = np.array([4,5])
#print(np.concatenate([a,b], axis = None))

# We're missing lengths 12 and 4
```
