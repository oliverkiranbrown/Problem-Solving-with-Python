```python
# Code to generate the matrix spiral
import numpy as np
'''
if N is odd:
    move right one step
    move down N steps
    move left N steps
else:
    move left one step
    move up N steps
    move right N steps
'''
# note that n must be odd!!
def spiral(n):
  g=np.zeros((n,n),dtype=int)
  r = int(n/2)
  c = int(n/2)
  count = 1
  g[r][c] = count
  count += 1
  for w in range(1,n+1):
    if w % 2 == 1 and count < n*n:
      c += 1
      g[r][c] = count
      count += 1
      for k in range(w):
        r += 1
        g[r][c] = count
        count += 1
      for k in range(w):
        c -= 1
        g[r][c] = count
        count += 1

    elif w % 2 == 0 and count < n*n:
      c -= 1
      g[r][c] = count
      count += 1
      for k in range(w):
        r -= 1
        g[r][c] = count
        count += 1
      for k in range(w):
        c+=1
        g[r][c] = count
        count += 1

  return g

spiral(13)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

def colour(rate,n):
  g = spiral(n)
  mapped = np.zeros((n,n))
  for i in range(n):
    for j in range(n):
      mapped[i][j] = g[i][j] % rate

  return mapped

print(colour(2,13))
plt.imshow(colour(2,13), cmap='hsv', vmin = 0, vmax = 2)
plt.show()
plt.axis('off')

print(colour(3,13))
plt.imshow(colour(3,13), cmap='hsv', vmin = 0, vmax = 3)
plt.show()
plt.axis('off')

plt.imshow(colour(4,13), cmap='hsv', vmin = 0, vmax = 4)
plt.show()
plt.axis('off')
```

```python
def create_map(n,rate):
  g = spiral(n)
  mapped = np.zeros((n,n))
  for i in range(n):
    for j in range(n):
      mapped[i][j] = g[i][j] % rate

  plt.imshow(colour(rate,n), cmap='hsv', vmin = 0, vmax = rate)
  plt.title(str(n)+'x'+str(n)+' spiral with '+str(rate)+' colours')
  plt.axis('off')
  plt.show()

create_map(13,3)
create_map(55,5)
create_map(255,5)
```

```python
# Let's check for primes!

def is_prime(n):
  for i in range(2,int(np.sqrt(n))+1):
    if (n%i) == 0:
      return 0
  return 1

def prime_spiral(n):
  g = spiral(n)
  mapped = np.zeros((n,n))
  for i in range(n):
    for j in range(n):
      mapped[i][j] = is_prime(g[i][j])

  plt.imshow(mapped, cmap='hsv', vmin = 0, vmax = 2)
  plt.title(str(n)+'x'+str(n)+' spiral with primes coloured')
  plt.axis('off')
  #plt.figure(figsize=(16, 16))
  plt.show()

prime_spiral(525)
```

```python
# Now we want to guess polynomials that cover primes and see how good they are...

def is_prime(n):
  for i in range(2,int(np.sqrt(n))+1):
    if (n%i) == 0:
      return False
  return True

def polynomial_spiral(n,p):
  # first entry counts primes, second entry counts polynomial crossovers
  prime_count = [0,0]
  g = spiral(n)
  mapped = np.zeros((n,n))
  for i in range(n):
    for j in range(n):
      # colour the primes
      prime = is_prime(g[i][j])
      if prime == True:
        prime_count[0] += 1
        mapped[i][j] = 1
      # colour the chosen polynomial
      for k in range(n):
        if g[i][j] == p(k):
          mapped[i][j] = 10
      if mapped[i][j] == 10 and prime == True:
        prime_count[1] += 1

  print("There were "+str(prime_count[0])+" primes and the polynomial found "+str(prime_count[1]))
  print("This makes "+str(np.round(prime_count[1]/prime_count[0]*100,3))+"% primes found by polynomial.")

  plt.imshow(mapped, cmap='Greys', vmin = 0, vmax = 2)
  plt.title(str(n)+'x'+str(n)+' spiral with primes coloured')
  plt.axis('off')
  plt.show()

polynomial_spiral(225,lambda n: n*n)
```
